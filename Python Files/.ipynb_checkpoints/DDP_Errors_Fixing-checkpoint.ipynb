```python
# Checking Orders and Customers consistency

Orders['customerID'] = Orders['customerID'].astype(int)
Customers['customerID'] = Customers['customerID'].astype(int)

missing_cust_mask = ~Orders['customerID'].isin(Customers['customerID'])
print("Orders with missing parent customerID:", missing_cust_mask.sum())

Orders[missing_cust_mask].head()
```

```python
# Checking Order_items and Products consistency

Order_Items['orderID'] = Order_Items['orderID'].astype(int)
Order_Items['productID'] = Order_Items['productID'].astype(int)

Orders_ids   = set(Orders['orderID'])
Products_ids = set(Products['productID'])

missing_order_mask   = ~Order_Items['orderID'].isin(Orders_ids)
missing_product_mask = ~Order_Items['productID'].isin(Products_ids)

print("Order_items with missing parent orderID:",   missing_order_mask.sum())
print("Order_items with missing parent productID:", missing_product_mask.sum())

Order_Items[missing_order_mask].head()
```
