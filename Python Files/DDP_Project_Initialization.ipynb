```python
# Rebuilding the Orders df for better results

import numpy as np
import pandas as pd

# Columns we care about (same as before)
ord_cols = [
    'Order Id',
    'order date (DateOrders)',
    'Order Status',
    'Order City',
    'Order State',
    'Order Country',
    'Order Zipcode',
    'shipping date (DateOrders)',
    'Shipping Mode',
    'Delivery Status',
    'Late_delivery_risk',
    'Days for shipping (real)',
    'Days for shipment (scheduled)',
    'Benefit per order',
    'Sales per customer',
    'Order Profit Per Order',
    'Customer Id'
]

Orders_raw = data[ord_cols].copy()

# 1. Minimal hard requirement: must have an Order Id and Customer Id
Orders_raw = Orders_raw.dropna(subset=['Order Id', 'Customer Id'])

# 2. Handle missing dates:
#    - Require order date (if it's missing, drop that row â€“ usually very few)
Orders_raw = Orders_raw.dropna(subset=['order date (DateOrders)'])

#    - For missing shipping date, fallback to order date (so NOT NULL is safe)
Orders_raw['shipping date (DateOrders)'] = Orders_raw['shipping date (DateOrders)'] \
    .fillna(Orders_raw['order date (DateOrders)'])

# 3. Fill string columns with "Unknown" instead of dropping
for col in ['Order Status', 'Order City', 'Order State',
            'Order Country', 'Order Zipcode',
            'Shipping Mode', 'Delivery Status']:
    Orders_raw[col] = Orders_raw[col].fillna('Unknown')

# 4. Fill numeric columns with 0 where missing
for col in ['Late_delivery_risk',
            'Days for shipping (real)',
            'Days for shipment (scheduled)',
            'Benefit per order',
            'Sales per customer',
            'Order Profit Per Order']:
    Orders_raw[col] = Orders_raw[col].fillna(0)

# 5. Cast to proper dtypes
Orders_raw['Order Id'] = Orders_raw['Order Id'].astype(int)
Orders_raw['Customer Id'] = Orders_raw['Customer Id'].astype(int)
Orders_raw['order date (DateOrders)'] = pd.to_datetime(Orders_raw['order date (DateOrders)'])
Orders_raw['shipping date (DateOrders)'] = pd.to_datetime(Orders_raw['shipping date (DateOrders)'])

# 6. One row per orderID:
#    If multiple rows exist (one per line item), keep the earliest record per order
Orders_for_db = (
    Orders_raw
    .sort_values('order date (DateOrders)')
    .drop_duplicates(subset=['Order Id'], keep='first')
    .copy()
)

print("Distinct orderIDs in Orders_for_db:", Orders_for_db['Order Id'].nunique())
print("Rows in Orders_for_db:", len(Orders_for_db))

# 7. Rename to match your physical model
Orders_for_db = Orders_for_db.rename(columns={
    'Order Id': 'orderID',
    'order date (DateOrders)': 'order_date',
    'Order Status': 'order_status',
    'Order City': 'order_city',
    'Order State': 'order_state',
    'Order Country': 'order_country',
    'Order Zipcode': 'order_zip',
    'shipping date (DateOrders)': 'shipping_date',
    'Shipping Mode': 'shipping_mode',
    'Delivery Status': 'delivery_status',
    'Late_delivery_risk': 'late_delivery_risk',
    'Days for shipping (real)': 'shipping_days_real',
    'Days for shipment (scheduled)': 'shipping_days_scheduled',
    'Benefit per order': 'benefit_per_order',
    'Sales per customer': 'sales_per_customer',
    'Order Profit Per Order': 'order_profit_per_order',
    'Customer Id': 'customerID'
})

# 8. Export this version for MySQL
Orders_for_db.to_csv('/Users/rohityadav/Desktop/Assignments/Sem 3/Database Design and Prototyping/Project/Dataset/csv Files/Orders.csv', index=False)
```
